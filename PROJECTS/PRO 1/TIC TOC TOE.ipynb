```python
import random
def choose_f():
    if random.randint(0, 1) == 0:
        return 'player2'
    else:
        return 'player1'
```

```python
def space_check(b,position):
    return b[position] == ''
    
```

```python
def full_check(b):
    for _ in range(1,10):
        if space_check(b,_):
            return False
    else:
        return True  
        
```

```python
def player_choice(b):
    position=0
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(b,position):
        position=int(input('CHOOSE A POSITION [1-9] '))
    return position    
```

```python
def replay():
    choice =input('WANNA PLAY AGAIN ? YES or NO').lower()
    return choice == 'yes'
```

```python
print('WELCOME TO TIC TOC TOE')
while True:
    #SET UP EVERYTHING.
    the_board=[' '] * 10
    player1_marker,player2_marker=player_input()
    turn=choose_f()
    print( turn + '  goes first !!!')
    play_game=input('Ready to play ? Y  or  N')
    if play_game == 'y':
        game_on=True
    else:
        game_on =False
     
    # GAME PLAY :
    while game_on:
        if turn == 'player1':
            # PLAYER 1 CHANCE.
            display_board(the_board)  # PLAYER CAN SEE THE BOARD.
            
            #Place the marker to its position.
            position=player_choice(the_board)
            place_marker(the_board,player1_marker,position)
            
            # check if it won.
            if win_check(the_board,player1_marker):
                display_board(the_board)
                print('PLAYER 1 YOU WON')
                game_on=False
                
            # If there is a tie possibilities.    
            else:
                if full_check(the_board):
                    display_board(the_board)
                    print('there is a tie!!!')
                    game_on=False
                else:
                    turn ='player2'
                    
        # PLAYER 2 TURNS !!!            
        else:
                        # PLAYER 1 CHANCE.
            display_board(the_board)  # PLAYER CAN SEE THE BOARD.
            
            #Place the marker to its position.
            position=player_choice(the_board)
            place_marker(the_board,player2_marker,position)
            
            # check if it won.
            if win_check(the_board,player2_marker):
                display_board(the_board)
                print('PLAYER 2 YOU WON')
                game_on=False
                
            # If there is a tie possibilities.    
            else:
                if full_check(the_board):
                    display_board(the_board)
                    print('THERE IS A TIE !!!')
                    game_on=False
                else:
                    turn = 'player1'
       
            
            
    
    
    
    
    if not replay():
        break
```
