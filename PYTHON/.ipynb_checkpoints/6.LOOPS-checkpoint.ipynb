```python
val=3
if (val == 1):
    print('DATA SCIENCE')
elif(val==2):
    print('MACHINE LEARNING')
else:
    print('NEURAL NETWORK')
```

```python
#Control flow: [for loop]
#Iterate through each object.eg.string,dictionary.
```

```python
course=['python','django']
for l in course:
    print(l)
```

```python
course=[1,2,3,4,'hello',5,6,7,8]
for l in course:
    print(l)
```

```python
c=[1,2,3,4,5,6,7,8]
for _ in c:
    if _%2==0:
        print(_)
```

```python

for _ in'hello world':
    print(_)
```

```python
list1=[(1,2),(3,4)] # TUPLES USES.
for _ in list1:
    print(_)
```

```python
for (a,b) in list1: #TUPLES UNPACKING
    print(a)
    print(b)
```

```python
d={'k1':1,'k2':2,'k3':3} #using dictionary.
for keys in d:
    print(keys)
```

```python
d={'k1':1,'k2':2,'k3':3} #using dictionary.
for keys in d.items():
    print(keys)
for keys in d.values():
    print(keys)    
```

```python
d={'k1':1,'k2':2,'k3':3} #unpacking
for keys,values in d:
    print(keys)
    print(values)
```

```python
#control flow: [while loops]
```

```python
x=0
while x<5:
    print(f'printed values:{x}')
    x=x+1
  
```

```python
#BREAK,PASS AND CONTINUE.
```

```python
f=[1,2,3]
for _ in f:
    pass
print('doesnt show error as we used pass ')
```

```python
d='saket prag'
for _ in d:
    if _ == 'a':
        continue  #continue skipped the letter a.
    print(_)   
```

```python
d='saket prag'
for _ in d:
    if _ == 'a':
        break # halt as soon as a appered.
    print(_)  
```
